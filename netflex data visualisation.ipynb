```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data=pd.read_csv(r"E:\data sets\data.csv")
```

```python
data.head(4)
```

```python
#count and plot the number of titles per genre
genre_count=data['genres'].value_counts()
genre_count.plot(kind='bar',title='Number of titles per genres')
```

```python
#find the average  IMDb rating by year
avg_rating_year=data.groupby('releaseYear')['imdbAverageRating'].mean()
avg_rating_year.plot(kind='line',title='the average  IMDb rating by year')
```

```python
#Display the top 10 titles based on IMDb rating.
top_rated=data.nlargest(10,'imdbAverageRating')[['title','imdbAverageRating']]
top_rated.plot(kind='barh',x='title',y='imdbAverageRating',title='top 10 heighest rated titles')
```

```python
#Number of Titles Released per Year
release_count=data['releaseYear'].value_counts().sort_index()
release_count.plot(kind='bar',title='number of titles released per year')
```
